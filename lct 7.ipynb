```python
#.............................. Standard Scaler...................
import pandas as pd
dataset={"Types_OF_Houses":["Flat","House","Flat","Flat","Flat","House","House","Flat","House","Flat","Flat","House","Flat","Flat"],"Size_of_House_BHK":[2,3,3,2,3,4,3,3,3,3,2,3,2,2],"Area_of_residence":[200,300,150,300,250,300,200,300,500,400,300,200,300,200],"prices_Lakhs":[13,12,140,15,16,12,13,124,15,16,102,13,15,16]}
housing_prices=pd.DataFrame(dataset)
print(housing_prices)
```

```python
stand=housing_prices["Area_of_residence"].std()
mean=housing_prices["Area_of_residence"].mean()
print(stand)
print(mean)
l=[]
for i in housing_prices["Area_of_residence"]:
    x=(i-mean)/(stand)
    l.append(x)
housing_prices["Normalized_Area"]=l
print(housing_prices)
```

```python
#checking of the Normalization
housing_prices["Normalized_Area"].std()
```

```python
housing_prices["Normalized_Area"].mean()
```

```python
st=housing_prices["prices_Lakhs"].std()
me=housing_prices["prices_Lakhs"].mean()
print(st)
print(me)
l=[]
for i in housing_prices["prices_Lakhs"]:
    x=(i-me)/(st)
    l.append(x)
housing_prices["Normalized_Prices"]=l
print(housing_prices)
```

```python
housing_prices["Normalized_Prices"].std()
```

```python
housing_prices["Normalized_Prices"].mean()
```

```python
9import matplotlib.pyplot as plt
plt.figure()
plt.scatter(housing_prices["Area_of_residence"],housing_prices["Normalized_Area"])
plt.xlabel("Actual_Area")
plt.ylabel("Normalized_Area")
plt.show()

```

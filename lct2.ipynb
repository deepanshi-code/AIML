```python
#*****MATPLOTLIB LIBRARY********
import matplotlib.pyplot as plt
import seaborn as  sns
import numpy as np
import pandas as pd
x = [1,2,3,4]
y = [10,20,25,30]
print("LINE GRAPH")
plt.plot(x, y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Graph")
plt.show()
```

```python
plt.figure()
print('BAR GRAPH')
names = ["A", "B", "C"]
marks = [80, 90, 85]
plt.bar(names, marks)
plt.show()
```

```python
plt.figure()
print("SCATTER PLOT")
plt.scatter(x, y)
plt.show()
```

```python
#**********SEABORN LIBRARY********
df = pd.DataFrame({"Day": [1,2,3,4],"Sales": [100,150,130,170]})
print("LINEPLOT")
sns.lineplot(x="Day", y="Sales", data=df)
plt.show()
```

```python
print("BARPLOT")
sns.barplot(x="Day", y="Sales", data=df)
plt.show()
```

```python
print("SCTATTERPLOT")
sns.scatterplot(x="Day",y="Sales",data=df)
plt.show()
```

```python
print("OUTLINERS")
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

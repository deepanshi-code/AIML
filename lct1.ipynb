```python
import numpy as np
print("******TYPE OF ARRAYS*******")
a = np.array([[1, 2, 3],[4,5,6]])
b=np.zeros(3)  
c=np.ones((2,3))    
d=np.arange(0,10,2) 
print(f"One dimension array :{a}\n Zero array :{b}\n One Array:{c}\n Ranged array:{d}\n")
```

```python
#array shape
print("*****OPERATION ON ARRAYS*****")
print(a.shape)
#dimension
print(a.ndim)
#size
print(a.size)
#dtype
print(a.dtype)
```

```python
#INDEXING
print("***INDEXING***")
print(a[0])
print(a[-1])
print(a[1:4])
```

```python
#***MATHS FUNCTIONS **

print("**MATHS FUNCTION**")
print(np.sqrt(a))      
print(np.mean(a))      
print(np.sum(a))
print(np.max(a))
print(np.min(a))
print(np.std(a))
```

```python
#************PADNDAS LIBRARY*****
import pandas as pd
data = {"Name": ["A", "B", "C"],"Marks": [80, 90, 85]}
df = pd.DataFrame(data)
print(df)
```

```python
print("***BASIC OPERATION***")
print(df.head())
print(df.tail())
print(df.shape)
print(df.info())
print(df.describe())

print("**INDEXING***\n")
print(df.loc[0])
print(df.iloc[0])
```

```python
print("**Filtering data**\n")
print(df[df["Marks"]>85])

print(df.drop_duplicates())
print(df.isnull())
print(df.isnull().sum())
print(df.fillna(0,inplace=True))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sales=pd.DataFrame({"Cust_Id":[101,102,103,104],"Cust_name":["ram","seeeta","geeta","riya"],"product":[["pen","eraser"],["maggie","cocacola"],["lipstick","eyeliner"],["pencil"]],"quantity":[[1,2],[5,3],[1,1],[1]],"price":[100,120,109,200],'category':["stationary","grocery","cosmetic","stationary"],"total_product":[2,2,2,1]})
print(sales)
```

```python
print(sales["Cust_name"])
```

```python
print(sales[["Cust_name","product"]])
```

```python
print(sales.isnull())
```

```python
sales.sort_values(by=["product"],ascending=False)
print(sales[['Cust_Id','Cust_name']].head())
```

```python
plt.figure()
plt.plot("Cust_name","total_product",data=sales)
plt.show()
plt.figure()
plt.bar("Cust_name","total_product",data=sales)
plt.figure()
```

```python
#****************************Deep First Search(DFS)*******************
grph={"A":["B","C"],"B":["E","D","A"],"C":["D","A","E"],"D":["F","C","B"],"E":["C","B","F"],"F":["D","E"]}
def dfs(st,goal,graph,path=[],vis=None):
    if vis is None:
        vis=set()
    vis.add(st)
    path.append(st)
    if st==goal:
        return path
    for neighbor in graph[st]:
        if neighbor not in vis:
            if dfs(neighbor,goal,graph,path,vis):
                return path
            return False
a=dfs("A","F",grph)
print(a)
```

```python
#*******BFS*******
from collections import deque
def bfs(graph,start,goal):
    de=deque([[start]])
    print("queue",de)
    vis=set([start])
    while de:
        path=de.popleft()
        print("path",path)
        node=path[-1]
        if node==goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in vis:
                vis.add(neighbor)
                new=path+[neighbor]
                de.append(new)
a=bfs(grph,'A','F')
print(a)
```
